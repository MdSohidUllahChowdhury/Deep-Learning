```python
from ultralytics import YOLO
import cv2
import cvzone #? pip install cvzone
import math

model = YOLO("/Users/home/development/Project/Own Deep Learning Project/yolo11x.pt")

cap = cv2.VideoCapture(0)  #? Try 0, 1, 2, etc. if 0 doesn't work
cap.set(3, 640)  # width
cap.set(4, 480)  # height

while True:
    success, img = cap.read()
    results = model(img, stream=True) #? stream=True to get results one by one (faster)
    for i in results:
        boxes = i.boxes
        for j in boxes:
            x1, y1, x2, y2 = j.xyxy[0] #? Get box coordinates

            x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2) #? Convert to integer
            w, h = x2 - x1, y2 - y1 #? Width and height
            cvzone.cornerRect(img, (x1, y1, w, h)) #? Draw rectangle with corners
            conf = math.ceil((j.conf[0] * 100)) / 100 #? Confidence
            print(conf)
            cvzone.putTextRect(img, f'{model.names[int(j.cls[0])]} {conf}', (max(0, x1), max(35, y1)), scale=1, thickness=1) #? Class name and confidence
            cv2.imshow("Webcam", img) #? Show webcam feed
            cv2.waitKey(1) #? 1 millisecond delay
```
